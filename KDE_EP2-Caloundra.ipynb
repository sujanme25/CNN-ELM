```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.special import ndtr
%matplotlib inline
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.distributions.mixture_rvs import mixture_rvs
from scipy.stats import zscore
import PIINTERVAL
import pandas as pd
import os
upper_bound = 0.975# 0.975/0.025(95),0.95/0.05(90),0.925/0.075(85),0.9/0.1(80),0.875/0.125(75),0.85/0.15(70)
lower_bound = 0.025
```

```python
df2=pd.DataFrame()
df2['TEST']=df1['YTEST']
df2['PRED']=df1['CELM']
df2['RES']=df1['YTEST']-df1['CELM']
df5=df2.copy()
muR=df5['RES'].mean()
sigR=df5['RES'].std()
df5['NR'] = zscore(df5['RES'])
kde = sm.nonparametric.KDEUnivariate(df5['NR'])
kde.fit(bw='silverman')
A1=kde.support
B1=kde.cdf
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111)
ax.plot(kde.support, kde.cdf, lw=3, label="CDF")
list = zip(B1, A1)
df = pd.DataFrame(list, columns=('cdf', 'R1'))
A=np.interp(lower_bound, df.cdf, df.R1)
B=np.interp(upper_bound, df.cdf, df.R1)
YU=df5['PRED']+(B*sigR+muR)
YL=df5['PRED']+(A*sigR+muR)
df5['LOWER']=YL
df5['UPPER']=YU
df_CELM=df5.copy()
y_test1=df_CELM['TEST'].values
y_test=df_CELM['TEST'].values
y_upper_TNET=df_CELM['UPPER'].values
y_lower_TNET= df_CELM['LOWER'].values
((y_test1 > y_upper_TNET).sum() + (y_test1 < y_lower_TNET).sum()) / y_test.shape[0]
pd.options.display.float_format = "{:,.3f}".format
act_range = max(y_test1)-min(y_test1)
confidence = 1 - (upper_bound - lower_bound)
PICP=PIINTERVAL.picp(y_test1, y_upper_TNET, y_lower_TNET)
MPIW = PIINTERVAL.mpiw( y_upper_TNET,y_lower_TNET)
WSSV=PIINTERVAL.WS(y_upper_TNET, y_lower_TNET, y_test1, confidence)
ARILV=PIINTERVAL.ARIL(y_test1, y_upper_TNET , y_lower_TNET)
PINAWV=PIINTERVAL.PINAW(act_range, y_upper_TNET , y_lower_TNET)
ACEV=PICP/100-(upper_bound-lower_bound)
F=(2*PICP/PINAWV)/(PICP+1/PINAWV)
PIM_TNET=[PICP,MPIW,WSSV,ARILV,PINAWV,ACEV,F]
PIM_TNET= pd.DataFrame(PIM_TNET).T
PIM_TNET.columns=['PICP','MPIW','WSSV','ARILV','PINAWV','ACEV','F']
dfPIM_CELM=pd.DataFrame(PIM_TNET)
dfPIM_CELM
```
